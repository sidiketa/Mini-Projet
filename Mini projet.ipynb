```python
import numpy as np
import pandas as pd
#from aspose import pycore
```

```python
# Importing the data
data_bfa = pd.read_table("BF.txt", delimiter="\t", header=None)
```

```python
data_bfa.head()
```

```python
# Defining columns names using the informations in the Readme.txt file 
data_bfa.columns = ["geonameid","name","asciiname","alternatenames", "latitude", "longitude","feature class",
                    "feature code","country code", "cc2","admin1 code","admin2 code","admin3 code", "admin4 code",
                    "population", "elevation", "dem", "timezone","modification date"]
```

```python
data_bfa.head()
```

```python
data_bfa.describe().T
```

```python
# Filtrage des données pour ne garder que les colonnes Identifiants, Noms de lieux, latitudes, longitudes
subset1_bfa = data_bfa[["geonameid","name", "latitude","longitude"]]
subset1_bfa.head()
```

```python
#Renommons les colonnes avec les noms suivants : 'ID', 'location_name', 'lat', 'long'
subset1_bfa.columns = ['ID', 'location_name', 'lat', 'long']
subset1_bfa.head()
```

```python
#Sauvegarde de ces données dans un fichier CSV, nommez-le burkina_location.csv
subset1_bfa.to_csv("burkina_location.csv", index= False)
```

```python
# Operations sur le fichier burkina_location
#burkina_location =pd.read_csv("burkina_location.csv")
burkina_location = subset1_bfa
```

```python
#extraction des données contenant le nom 'gounghin'
gounghin = burkina_location[burkina_location.location_name =='Gounghin']
gounghin
```

```python
#, sauvegarde des données gounghin sous le fichier gounghin.csv
gounghin.to_csv("gounghin.csv", index= False)
```

```python
# Données dont les noms les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique)
burkina_location=burkina_location.sort_values(by= "location_name", ascending= True)
```

```python
burkina_location.tail()
```

```python
# Remplacer les noms de localités par des noms sans accent en debut de mot
burkina_location.replace(to_replace= ["Èrmos","Èrza","Éléoui"], value=["Ermos","Erza","Eléoui"], inplace= True)
```

```python
# Refaire le tri par ordre alphabétique
burkina_location=burkina_location.sort_values(by= "location_name", ascending= True)
burkina_location
```

```python
# Données dont les noms des premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique
subset_A_to_P = burkina_location[burkina_location['location_name'].str[0].between('A', 'P', inclusive='both')]
subset_A_to_P
```

```python
# Identifions  la latitude minimale  et le nom de lieu correspondant
subset_lat_min = burkina_location[burkina_location["lat"] == min(burkina_location['lat'])]
subset_lat_min
```

```python
# Identifions la longitude minimale  et le nom de lieu correspondant
subset_long_min = burkina_location[burkina_location["long"] == min(burkina_location['long'])]
subset_long_min
```

```python
#les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)
subset_lat_lon = burkina_location[burkina_location["lat"] >= 11]
subset_lat_lon = subset_lat_lon[subset_lat_lon['long']<= 0.5]
subset_lat_lon
```

```python
#Création du fichier Excel mini_projet et des  feuilles  gounghin et A_to_P contenant les données de gounghin et subset_A_to_P
with pd.ExcelWriter('mini_projet.xlsx') as writer:  
    gounghin.to_excel(writer, sheet_name='gounghin')
    subset_A_to_P.to_excel(writer, sheet_name='A_to_P')
```
